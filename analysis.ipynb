```python
# Column Âge

for col in data.columns:
    if col.__contains__("âge"):
        data[col].replace('38 ans', '38', inplace=True)
        data[col].replace('23ans', '23', inplace=True)
        data[col].replace('27ans', '27', inplace=True)
        data[col].replace('18 ans', '18', inplace=True)
        data[col].replace('19 ans', '19', inplace=True)
        data[col].replace('19ans', '19', inplace=True)
        data[col].replace('21 ans', '21', inplace=True)
        data[col].replace('20 ans', '20', inplace=True)
        data[col].replace('49 سنة', '49', inplace=True)

        data[col] = pd.to_numeric(data[col])
        break

data.dtypes
```

```python
# Column 7

data['7.A quel âge avez-vous commencé à fumer?'].replace('/', np.nan, inplace=True)
data['7.A quel âge avez-vous commencé à fumer?'].replace('لاادخن ', np.nan, inplace=True)
data['7.A quel âge avez-vous commencé à fumer?'].replace('Jamais', np.nan, inplace=True)
data['7.A quel âge avez-vous commencé à fumer?'].replace('non', np.nan, inplace=True)
data['7.A quel âge avez-vous commencé à fumer?'].replace('12 ans', '12', inplace=True)

data['7.A quel âge avez-vous commencé à fumer?'] = pd.to_numeric(data['7.A quel âge avez-vous commencé à fumer?'])
```
